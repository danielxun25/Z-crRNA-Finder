```python
def find_seq(nt: int, input_seq: str, frame: int = 21, special_char: str = 'A'):
  result = []
  input_seq = input_seq.upper()
  for i in range(0, len(input_seq) - frame +1):
    cur = input_seq[i:i+frame]
    pam = input_seq[i-4:i]
    if cur.count(special_char) == nt and pam.startswith('TTT'):
      result.append(cur)
  if len(result) == 0:
    print('Nothing found')
  else:
    print(result)
```

```python
input_seq = ''

find_seq(12, input_seq)
```
